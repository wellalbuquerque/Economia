Aplicando-se a fórmula do Black-Scholes-Merton resolvemos o exercício, lembre-se que o ativo-objeto variou +1,00 então soma-se tal valor ao valor inicial do ativo-objeto (S).

```python
#Dados iniciais do exercício
S = (float(input("Qual o valor do ativo-objeto (S) negociado em R$: "))) 
K = (float(input("Qual o valor do Stike (K) em R$: "))) 
r = (float(input("Qual a taxa de juros a.a: "))) 
sigma =(float(input("Qual o desvio padrão: "))) 
p_Meses= (float(input("Qual a vigência em meses:")))
T = p_Meses/12
d1 = (np.log(S/K) + (r + (sigma**2) / 2)*T) / (sigma*np.sqrt(T))
d2 = d1 - sigma*np.sqrt(T)
Nd1 = norm.cdf(d1)
Nd2 = norm.cdf(d2)
C = S*Nd1 - K*np.exp(-r*T)*Nd2
print ('O valor do CALL corresponde a R$',"%.2f"% C, )
```
